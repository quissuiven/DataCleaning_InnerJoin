Goal: identify list of people who attended both Wednesday and Thursday Sessions

Data format: two different tables in Excel workbook

Solution: Using Python Pandas to clean and perform inner join on "ID"

## Cleaning Data

```python
import pandas as pd
```

```python
df = pd.read_excel('Test.xlsx',sheet_name='Exercise #4')
df
```

```python
df = df.drop(['Unnamed: 0','Unnamed: 5', 'Unnamed: 6','Unnamed: 7'],axis=1)   #drop columns, axis = 1
df = df.drop([0,1],axis = 0)            #drop row 0
df = df.rename(columns={"Wednesday Session": "Name", "Unnamed: 2": "ID", "Unnamed: 3": "Region", "Unnamed: 4": "Retake?", "Thursday Session": "Name", "Unnamed: 9": "ID", "Unnamed: 10": "Status","Unnamed: 11": "Retake?"})
```

```python
df1 = df.iloc[:, 0:3]                                          #splitting dataframe into 2
df2 = df.iloc[:, 4:7]
```

## Inner Join

```python
df_clean = pd.merge(df1, df2, on =['ID'], how='inner')         #other joins are: outer,left,right
```

```python
df_clean
```

```python
df_clean = df_clean.drop(['Name_y'],axis=1)   
df_clean = df_clean.rename(columns={"Name_x": "Name"})
df_clean
```
